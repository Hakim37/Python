## Merge /Join

```python
import numpy as np
import pandas as pd
```

```python
#concant or append operation

data ={
       'emp_id':['1','2','3','4','5'],
       'first_name':['Jason','Andy','Anllen','Alice','Amy'],
       'last_name' :['Larkin','Jacob','A','AA','Jackson']}
```

```python
df1=pd.DataFrame(data, columns = ['emp_id','first_name','last_name'])
df1
```

```python
data={
    'emp_id':['4','5','6','7'],
    'first_name':['Brain','Shize','Kim','Jose'],
    'last_name':['Alexander','Suma','Mike','G']}
```

```python
df2 = pd.DataFrame(data,columns=['emp_id','first_name','last_name'])
df2
```

```python
#Step2-Introduction to machine learning
```

```python
#Uningconcat
df=pd.concat([df1,df2])
df
```

```python
# OR
# Using append

df1.append(df2)
```

### Join the two dataframes along columns

```python
pd.concat([df1,df2],axis=1)
```

### Merge two dataframes

```python
# Merge two data frames based on the emp_id values
# in this case only the emp_id's present in both table will be joined

pd.merge(df1,df2, on='emp_id')
```

### Join

```python
#Pandas offer SQL Stule merges ges as well
```

### left join

```python
pd.merge(df1,df2,on='emp_id',how='left')
```

```python
# Merge while adding a suffix to duplicate column name of both table
pd.merge(df1,df2,on='emp_id',how='left',suffixes=('_left','right'))
```

# step2 -introduction to machine learning

```python
# Right jion two data Frame
#Right

pd.merge(df1,df2, on='emp_id',how='right')
```

## Inner join two dataframes

```python

pd.merge(df1,df2,on='emp_id',how ='inner')
```

## Outer join two data frame

```python
pd.merge(df1,df2,on='emp_id',how='outer')
```
